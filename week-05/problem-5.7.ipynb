## (a) __Histogram of $1/x$__

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = default_rng()
rand = rng.random
```

```python
from random import seed
from random import random

t = []

for _ in range(10000):
    value = random()
    a = 1/value
    t.append(a)
    #print(a)
    
#print(t)
```

```python
plt.hist(t, bins=10000)
plt.xlim([0, 100])
```

## (b) Log-log plot of frequencies

```python
plt.hist(t, bins=10000)
plt.loglog()
plt.xlim([1, 100])
```

## (c) Analysis and Conclusions

The distobution is logarithmic which makes sense as the change of the function is exponetionial. We know that there is a relationship between 1/x and ln(x) from calculus class.
