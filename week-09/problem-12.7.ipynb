You should choose at least 4 pairs of initial concentrations $[\bar{c}_1(0), \bar{c}_2(0)]$ to illustrate what happens in all 4 quadrants of your phase portrait.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import factorial
from numpy.random import default_rng
from scipy.signal import convolve
rand = default_rng().random
from scipy.integrate import odeint

R=1

def f(Y, t):
    y1, y2 = Y
    
    return [R*y1*y2, y1*(R*y2-1)]

y1 = np.linspace(-4.0, 4.0, 20)
y2 = np.linspace(-5.0, 5.0, 20)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
        
        
        
for y00 in [0, 0.5, 1, 1.5, 2]:
    span0 = np.linspace(0, 50, 200)
    y0 = [6.0, y00]
    ys0 = odeint(f, y0, span0)
        
for y01 in [0, 0.5, 1, 1.5, 2]:
    span1 = np.linspace(0, 50, 200)
    y1 = [7.0, y01]
    ys1 = odeint(f, y1, span1)
    
for y02 in [0, -0.5, -1, -1.5, -2]:
    span2 = np.linspace(0, 50, 200)
    y2 = [6.0, y02]
    ys2 = odeint(f, y2, span2)    
        
for y03 in [0, -0.5, -1, -1.5]:
    span3 = np.linspace(0, 50, 200)
    y3 = [8.5, y03]
    ys3 = odeint(f, y3, span3)      
    

plt.figure(dpi=200)
plt.xlim([-5, 10])    
plt.ylim([-10,5])
        
Q = plt.quiver(Y1, Y2, u, v, color='r')
plt.plot(ys0[:,0], ys0[:,1], 'b-')
plt.plot([ys0[0,0]], [ys0[0,1]], 'o')

plt.plot(ys1[:,0], ys1[:,1], 'b-')
plt.plot([ys1[0,0]], [ys1[0,1]], 'o') 

plt.plot(ys2[:,0], ys2[:,1], 'b-')
plt.plot([ys2[0,0]], [ys2[0,1]], 'o') 


plt.plot(ys3[:,0], ys3[:,1], 'b-') 
plt.plot([ys3[0,0]], [ys3[0,1]], 'o')
```
