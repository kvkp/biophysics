```python
import numpy as np
import matplotlib.pyplot as plt
def walk1(n):
   x1 = 0
   y1 = 0
   xposition = [0] 
   yposition = [0] 
   for i in range (1,n+1):
       step = np.random.uniform(0,1)
       if step < 0.5:     
           x1 += 1
           y1 += -1  
       if step > 0.5:  
           x1 += -1
           y1 += 1 
       if step < 0.25:     
           x1 += 1
           y1 += 1  
       if step > 0.75:  
           x1 += -1
           y1 += -1 



       xposition.append(x1)
       yposition.append(y1)
   return [xposition,yposition]


Randwalk1 = walk1(1000) 
plt.plot(Randwalk1[0],Randwalk1[1],'r-', label = "Randwalk1D") 

Randwalk2 = walk1(1000) 
plt.plot(Randwalk2[0],Randwalk2[1],'g-', label = "Randwalk1D")

Randwalk3 = walk1(1000) 
plt.plot(Randwalk3[0],Randwalk3[1],'b-', label = "Randwalk1D") 

Randwalk4 = walk1(1000) 
plt.plot(Randwalk4[0],Randwalk4[1],'y-', label = "Randwalk1D") 

plt.title("Random Walks")
plt.show()
```
